#### Filter rows of a pandas DataFrame by column value 

```python
boolean = []
for x in movies.duration:
    boolean.append(False) if x <200 else boolean.append(True)
```

```python
is_long = pd.Series(boolean)
```

```python
movies[is_long].head()
```

#### Short-way for filtering 

```python
is_long = movies.duration >=200
```

```python
movies[is_long].head()
```

#### yet another way

```python
movies[movies.duration>=200].head()
```
